```python
import pandas as pd

# Загружаем файлы
df1 = pd.read_csv('протоколы_проверка.csv', sep=';', encoding='utf-8')
df2 = pd.read_csv('Реестр_обученных.csv', sep=';', encoding='utf-8')

# Удаляем строки с пустыми ключевыми данными
df1 = df1.dropna(subset=['СНИЛС', 'Фамилия', 'Имя'])
df2 = df2.dropna(subset=['СНИЛС', 'Фамилия', 'Имя'])

# Поля для ключа
key_fields1 = ['Фамилия', 'Имя', 'Отчество', 'СНИЛС', 'Должность',
               'Номер протокола', 'Дата проверки знаний', 'Программа обучения']
key_fields2 = ['Фамилия', 'Имя', 'Отчество', 'СНИЛС', 'Должность',
               'Номер протокола', 'Дата проверки знаний', 'Программа обучения']

# 🔥 ГЛАВНОЕ: ОЧИЩАЕМ ДАННЫЕ (убираем пробелы, приводим к нижнему регистру)
for col in key_fields1:
    if col in df1.columns:
        df1[col] = df1[col].astype(str).str.strip().str.lower()

for col in key_fields2:
    if col in df2.columns:
        df2[col] = df2[col].astype(str).str.strip().str.lower()

# Проверяем, что поля для ключа существуют
missing1 = [col for col in key_fields1 if col not in df1.columns]
missing2 = [col for col in key_fields2 if col not in df2.columns]
if missing1:
    print(f"❌ В первом файле нет столбцов: {missing1}")
if missing2:
    print(f"❌ Во втором файле нет столбцов: {missing2}")

# Создаём ключи
df1['key'] = df1[key_fields1].astype(str).agg('|'.join, axis=1)
df2['key'] = df2[key_fields2].astype(str).agg('|'.join, axis=1)

# Находим столбец с регистрационным номером
reg_column = None
for col in df2.columns:
    if 'регистр' in col.lower() or 'номер' in col.lower():
        reg_column = col
        break

if reg_column is None:
    reg_column = df2.columns[0]
    print(f"\n⚠️ Использую первый столбец как регистрационный номер: '{reg_column}'")
else:
    print(f"\n✅ Найден столбец с регистрационным номером: '{reg_column}'")

# Объединяем таблицы
result = df1.merge(df2[['key', reg_column]], on='key', how='left')

# Убираем .0 (сохраняем как текст)
result[reg_column] = result[reg_column].fillna('').astype(str)
result[reg_column] = result[reg_column].str.replace(r'\.0$', '', regex=True)
result[reg_column] = result[reg_column].replace('', pd.NA)

# Переименовываем
result = result.rename(columns={reg_column: 'Регистрационный номер'})

# Проверяем результат
not_found = result['Регистрационный номер'].isna().sum()
total = len(df1)
print(f"\n📊 Всего строк в первом файле после очистки: {total}")
print(f"❌ Не найдено соответствий: {not_found}")
print(f"✅ Найдено соответствий: {total - not_found}")

# Показываем примеры найденных и ненайденных
print("\n📋 Примеры НАЙДЕННЫХ регистрационных номеров (первые 5):")
found_examples = result[result['Регистрационный номер'].notna()][['Фамилия', 'Имя', 'Программа обучения', 'Регистрационный номер']].head(5)
print(found_examples)

print("\n🔍 Примеры НЕ НАЙДЕННЫХ (первые 5):")
not_found_examples = result[result['Регистрационный номер'].isna()][['Фамилия', 'Имя', 'Программа обучения']].head(5)
print(not_found_examples)

# Сохраняем результат
result.to_csv('результат.csv', sep=';', index=False, encoding='utf-8-sig')
print("\n✅ Готово! Файл 'результат.csv' создан.")
# Показываем подробную информацию о несовпадающих строках
print("\n🔍 ДЕТАЛЬНЫЙ АНАЛИЗ НЕСОВПАДЕНИЙ:")
not_found_df = result[result['Регистрационный номер'].isna()]

if len(not_found_df) > 0:
    # Берем первую несовпадающую строку
    example = not_found_df.iloc[0]
    print("\nПример несовпадающей строки из первого файла:")
    for field in key_fields1:
        print(f"  {field}: {example[field]}")

    # Ищем её во втором файле без приведения к нижнему регистру
    print("\n🔎 Поиск во втором файле (оригинальные данные):")
    # Загружаем оригинальные данные без очистки
    df2_orig = pd.read_csv('Реестр_обученных.csv', sep=';', encoding='utf-8')
    # Ищем по СНИЛС и программе
    snils = example['СНИЛС']
    prog = example['Программа обучения']
    found = df2_orig[(df2_orig['СНИЛС'].astype(str).str.strip().str.lower() == snils) &
                     (df2_orig['Программа обучения'].astype(str).str.strip().str.lower() == prog)]
    if len(found) > 0:
        print("Найдено во втором файле:")
        print(found[['Фамилия', 'Имя', 'Должность', 'Программа обучения']].head(1))
    else:
        print("Не найдено по СНИЛС и программе")

```
